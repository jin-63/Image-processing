```python
# %%writefile은 순수 텍스트 파일로 저장해줍니다.
%%writefile README.md
** 컴퓨터 개념이란 **
 내장된 지능과 호환 가능하도록 컴퓨터가 정지된 영상(Image) 또는 동영상에서 특징(Feature)을 찾아내고, 그 의미를 사람처럼 이해하도록 프로그램하는 것을 목적

** 이미지 전처리(Preprocessing) **
이미지의 특징을 추출하고 추적할 수 있도록 원하는 목적에 따라 이미지를 리뷰하는 과정

** 특징 추출 및 추적(Feature Extraction and Tracking) **
정확하고 믿을 수 있는 이미지의 분석 결과를 위해 다양한 변수 처리자를 사용하여

## ** 이미지 프로세싱1 **
- Google 코랩에서 OpenCV를 사용하여 이미지 출력 및 Numpy배열을 표현 및 이름의 수 확인
- matplotlib는 RGB를 사용하지만 OpenCV는 BGR을 사용하여 서로 간의 차이를 확인합니다.
- ROI(Region of Interest)는 이미지나 영상에서 '관심있는 지역'을 지정합니다.
- 고양이 이미지의 얼굴 부분 ROI로 선거
- 사모 지붕 이미지의 BGR채널 분리하고 출력, 분리한 BGR을 다시 생각하는 RGB, BGR로 자체 표현
- 분리한 이미지의 RGB를 각각의 색상 값으로 표현해서 출력하고 다른 색상으로 변환하기
- canvas생성 후 OpenCV로 그림그림기(선, 도형, 쓰기)
- 방금 만든 고양이 이미지 ROI부분에 CAT ROI 표현하기

## ** 이미지프로세싱2 **

### 스레시홀딩(Thresholding, 청년프로세스)
경계점 기준으로 이미지 속을 이루는 모양과 경계선을 이해하기 쉽게 이해하기 위한 바이너리 이미지를 사용하기

- retVal , result = cv2.threshold(sourceImage, thresholdVal, maxVal, thresholdType)
- retVal -> 스레시홀더에 사용되는 경계값을 돌려드립니다
- sourceImage -> 스레시홀딩을 할 이미지
- threshVal -> 스레시홀더에 사용되는 경계값
- maxVal -> 경계값을 초과할 때 매기는 값
- ThresholdType -> 스레시 홀딩 타입

### 이미지 연산(OpenCV와 NumPy연산의 구별)
두 가지 이미지를 서로 빼거나 더하는것으로 새로운 결과 이미지를 생성합니다(서로 곱하거나 나누지 않음)

** OpenCV 사용 **
- 클리핑(clipping): OpenCV는 광대역 범위 값이 255를 초과하면 값을 클리핑(clipping)해서 255를 유지합니다.
- cv2.add(np.uint8( [ 200 ] ), np.uint8( [ 100 ] )) 는 200+100= 300이지만 255가 최대
- cv2.subtract(np.uint8( [ 50 ] ), np.uint8( [ 100 ] )) 은 50-100= -50이지만 0이 최소
** Numpy 사용 **
- 랩어라운드(wrap around): NumPy는 225가 많은 경우 공유 값을 0부터 다시 더 많은 범위(wrap around) 때문에 255을 공유 거리에서 1을 더 뺌(0부터 시작)
- np.uint8( [ 200 ] ) + np.uint8( [ 100 ] ) 은 200+100= 300이지만 300-256=44가 최대
- np.uint8( [ 50 ] ) - np.uint8( [ 100 ] )은 50-100= -50이지만 256-50=206이 최소

### 비트와이즈 배열(bitwise-option)
두 이미지를 혼합할 때 특정 지역만 또는 특정 범위만 제외하는 등 선택적인 공유에 사용
두 이미지는 같은 모양을 가지고 있어야 합니다.
- AND ->찾는 두 쌍이 모두 0보다 크면 ture로 돌아와요
- 또는 -> 비교하는 두 개의 낮은 평행 배열이 0보다 크거나 둘 사이 0보다 크면 true로 돌아갑니다
- XOR -> 비교하는 두 개의 확장된 확장형 배열이 0보다 넓은 두 개의 확장형과 동시에 0보다 유연할 때 true를 반환합니다.
- NOT-> 하나의 이미지가 0이면 true이고, 0이 아니면 false를 반환합니다(바이너리 이미지 반전 효과)

### 이미지 마스킹(Image Masking)
하나의 이미지에서 원하는 영역만 분리하는 작업
- 마스크(mask)는 바이너리 이미지로, 특정 이미지에서 0이 아닌 슬라이더만 올라갈 수 있도록 함

### HSV 컬러 범위로 스레시홀딩 후 이미지 마스킹하기
RGB보다 HSV컬러채널이 더 색범위를 구하기 쉬움
OpenCV는 uint8 자료형으로 0부터 255 값으로 색상과 채도, 명도 값을 표현
- Hue(색상)은 360°/2=180°로 [0,179] 사이 값으로 표현하고 180-255까지의 값은 179로 표현
- Saturation(채도)는 0-100% 값을 [0,255]범위로 표현
- Value(명도)는 0-100% 값을 [0,255]범위로 표현
```
